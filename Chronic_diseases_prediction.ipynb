Preprocessing of data

Assigning feature names

```python
data.columns = ["age","blood_pressure","specific_gravity","albumin","sugar","red_blood_cells","pus_cells","pus_cell_clumps","bacteria","blood_glucose_random","blood_urea","serum_creatinine","sodium","potassium","haemoglobin","packed_cell_volume","white_blood_cell_count","red_blood_cell_count","hypertension","diabetes_mellitus","coronory_artery_disease","appetite","peda_edema","anemia","class"]   
```

Converting data types

```python
text_columns = ["packed_cell_volume","white_blood_cell_count","red_blood_cell_count"]
for i in text_columns:
    print(i,":", data[i].dtype)
```

```python
def convert_text_numeric(data,column):
    data[column] = pd.to_numeric(data[column],errors='coerce') # invalid parsing will set as NAN
    
for j in text_columns:
    convert_text_numeric(data,j)
    print(f"{j} : {data[j].dtype}")
```

```python
data.info()
```

Handling missing values

```python
missing = data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
def mean_imputation(data,column):
    mean_value = data[column].mean()
    data[column].fillna(value = mean_value,inplace=True)
    
def mode_imputation(data,column):
    mode = data[column].mode()[0] # producesa series of mode 
    data[column].fillna(mode,inplace = True)
```

```python
num_columns = [ col for col in data.columns if data[col].dtype != 'object']
for col_names in num_columns:
    mean_imputation(data,col_names)
```

```python
cat_columns = [col for col in data.columns if data[col].dtype == 'object']
for col_name in cat_columns:
    mode_imputation(data,col_name)
```

```python
missing = data.isnull().sum()
missing[missing>0].sort_values(ascending=False).head(20)
```

```python
data.head(5)
```

Checking the number of categories in a column 

```python
print(f"dibetes_mellitus : {data['diabetes_mellitus'].unique()}")
print(f"coronory_artery_disease : {data['coronory_artery_disease'].unique()}")
print(f"class : {data['class'].unique()}")
```

```python
data["diabetes_mellitus"] = data["diabetes_mellitus"].replace(to_replace={" yes":"yes","\tno":"no","\tyes":"yes"})
data['coronory_artery_disease'] = data['coronory_artery_disease'].replace(to_replace={"\tno":"no"})
data["class"] = data["class"].replace(to_replace={"ckd\t":"ckd"})
```

```python
print(f"dibetes_mellitus : {data['diabetes_mellitus'].unique()}")
print(f"coronory_artery_disease : {data['coronory_artery_disease'].unique()}")
print(f"class : {data['class'].unique()}")
```

Feature encoding

```python
data['class'] = data['class'].map({'ckd':1,'notckd':0})
data['red_blood_cells'] = data['red_blood_cells'].map({'normal':1,'abnormal':0})
data['pus_cells'] = data['pus_cells'].map({'normal':1,'abnormal':0})
data['pus_cell_clumps'] = data['pus_cell_clumps'].map({'present':1,'notpresent':0})
data['bacteria'] = data['bacteria'].map({'present':1,'notpresent':0})
data['hypertension'] = data['hypertension'].map({'yes':1,'no':0})
data['diabetes_mellitus'] = data['diabetes_mellitus'].map({'yes':1,'no':0})
data['coronory_artery_disease'] = data['coronory_artery_disease'].map({'yes':1,'no':0})
data['appetite'] = data['appetite'].map({'good':1,'poor':0})
data['peda_edema'] = data['peda_edema'].map({'yes':1,'no':0})
data['anemia'] = data['anemia'].map({'yes':1,'no':0})
```

Splitting training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
data['class'].value_counts()
```

```python
x = data.drop('class',axis = 1)
y = data['class']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.25,random_state = 25)
```

```python
print(f"x train shape : {x_train.shape}")
print(f"x test shape : {x_test.shape}")
print(f"y train shape : {y_train.shape}")
print(f"y test shape : {y_test.shape}")
```

Using a single Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dsc = DecisionTreeClassifier()
dsc.fit(x_train,y_train)
```

```python
y_pred = dsc.predict(x_test)
y_pred
```

Applying Algorithms

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
models =[]
models.append(('naive_bayes',GaussianNB()))
models.append(('random_forest',RandomForestClassifier()))
models.append(('svm',SVC(kernel='linear')))  
models.append(('decision_tree',DecisionTreeClassifier()))
#models.append(('k_means',KNeighborsClassifier(n_neighbors=2)))
```

```python
models
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
for name , model in models:
    print(name,model)
    model.fit(x_train,y_train)
    y_predict = model.predict(x_test)
    print("confusion : ",confusion_matrix(y_test,y_predict))
    print("accuracy : ",accuracy_score(y_test,y_predict))
    print("precision : ",precision_score(y_test,y_predict))
    print("recall : ",recall_score(y_test,y_predict))
    print("f1 : ",f1_score(y_test,y_predict))
    print('\n')
```
